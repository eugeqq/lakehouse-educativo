# Notebook 1 — Ingesta Bronze
**Proyecto:** Lakehouse Educativo  
**Capa:** Bronze — datos crudos sin transformación  
**Fuente:** CSVs subidos al Volume `/Volumes/main/edu_lakehouse/raw_files/`  
**Objetivo:** Persistir los datos en su estado original como tablas Delta para garantizar trazabilidad.




## Configuración inicial
Creamos el schema `edu_lakehouse` dentro del catalog `main` para organizar todas las tablas del proyecto.

```python
spark.sql("CREATE SCHEMA IF NOT EXISTS main.edu_lakehouse")
spark.sql("USE main.edu_lakehouse")
print("Schema listo")
```

## Bronze — Alumnos (exploración)
Leemos el CSV raw sin modificaciones. El objetivo es ver el estado original del dato antes de cualquier transformación.

```python
df_alumnos = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv("/Volumes/main/edu_lakehouse/raw_files/alumnos.csv")
)

df_alumnos.printSchema()
df_alumnos.show(20, truncate=False)
print(f"Filas cargadas: {df_alumnos.count()}")
```

## Bronze — Alumnos (persistencia)
Guardamos el dato crudo como tabla Delta. En Bronze no limpiamos nada, conservamos el origen para trazabilidad y auditoría.

```python
(df_alumnos
    .write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("main.edu_lakehouse.bronze_alumnos")
)
print("✓ bronze_alumnos creada")
```
