```python
import subprocess
import os

# Crear schema y volume
spark.sql("CREATE CATALOG IF NOT EXISTS main")
spark.sql("CREATE SCHEMA IF NOT EXISTS main.edu_lakehouse")
spark.sql("CREATE SCHEMA IF NOT EXISTS main.edu_quarantine")
spark.sql("""
    CREATE VOLUME IF NOT EXISTS main.edu_lakehouse.raw_files
""")
print("✓ Schema y Volume creados")
```

```python
# Ruta del repo en Databricks
repo_path = "/Workspace/Repos/quadromariaeugenia@uca.edu.ar/lakehouse-educativo/data"

# Copiar CSVs del repo al Volume
archivos = ["alumnos.csv", "materias.csv", "asistencia.csv", "notas.csv"]

for archivo in archivos:
    origen = f"{repo_path}/{archivo}"
    destino = f"/Volumes/main/edu_lakehouse/raw_files/{archivo}"
    dbutils.fs.cp(f"file:{origen}", destino)
    print(f"✓ {archivo} copiado al Volume")

print("\nArchivos en el Volume:")
display(dbutils.fs.ls("/Volumes/main/edu_lakehouse/raw_files/"))
```

```python
# Verificar que todo está listo
print("=== Schemas creados ===")
spark.sql("SHOW SCHEMAS IN main").show()

print("\n=== Archivos en Volume ===")
display(dbutils.fs.ls("/Volumes/main/edu_lakehouse/raw_files/"))

print("\n✓ Setup completo. Podés ejecutar los notebooks en orden:")
print("  01_bronze_ingesta")
print("  02_silver_limpieza_v2")
print("  03_gold_indicadores")
```
